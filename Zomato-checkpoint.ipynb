#### Now we will join the country column from the excel file  with the csv dataset

```python
df_final=pd.merge(df,df1,on="Country Code",how ="left")
df_final
```

```python
Rating_count=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
Rating_count['Rating color'] = Rating_count['Rating color'].replace(["White","Dark Green"],["Blue","Pink"]) 
Rating_count
```

<!-- when the Rating is in between 4.5 to 4.9--- it is excellent.
when the Rating is in between 4.0 to 4.4--- it is very good
when the Rating is in between 3.5 to 3.9--- it is good
when the Rating is in between 2.5 to 3.4--- it is Average.
when the Rating is in between 1.8 to 2.4--- it is poor. -->

```python
plt.figure(figsize=(20,10))
d1=sns.barplot(x="Aggregate rating",y="Rating Count",data=Rating_count,saturation=1)
for bars in d1.containers:
    d1.bar_label(bars)
```

```python
#countplot (Frequency of colours)
sns.countplot(x="Rating color",data=Rating_count,palette=['Blue','Red','Orange','Yellow','Green','Pink'])
```
